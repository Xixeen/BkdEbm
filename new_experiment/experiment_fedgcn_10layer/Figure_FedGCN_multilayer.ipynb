```python
import collections
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_data(file_name):
    a = open(file_name, 'r')
    train_loss = collections.defaultdict(float)
    train_acc = collections.defaultdict(float)
    val_loss = collections.defaultdict(float)
    val_acc = collections.defaultdict(float)
    test_loss = 0
    test_acc = 0
    count = 0
    for line in a:
        line = line.split()
        if line[1] == 'train':
            train_loss[int(line[0])] += float(line[2])
            train_acc[int(line[0])] += float(line[3])
        elif line[1] == 'val':
            val_loss[int(line[0])] += float(line[2])
            val_acc[int(line[0])] += float(line[3])
        elif line[1] == 'test':
            test_loss += float(line[2])
            test_acc += float(line[3])
            count += 1
        else:
            print("error")
    a.close()
    for key in train_loss.keys():
        train_loss[key] /= count
        train_acc[key] /= count
        val_loss[key] /= count
        val_acc[key]  /= count
    test_loss /= count
    test_acc /= count
    
    return train_loss, train_acc, val_loss, val_acc, test_loss, test_acc
```

```python
def get_plot(file_name):
    train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = get_data(file_name)
    
    #plt.plot(train_loss.keys(), train_loss.values(), label = 'train_loss')
    #print(train_acc.values())
    plt.plot(train_acc.keys(), train_acc.values(), label = 'train_acc')
    #plt.plot(val_loss.keys(), val_loss.values(), label = 'val_loss')
    plt.plot(val_acc.keys(), val_acc.values(), label = 'val_acc')
    plt.ylim(0, 1) 
    plt.xlim(0, 300) 
    plt.legend()
    plt.show()
```

```python
plt.style.use('seaborn-whitegrid')

test_acc_list = []
train_acc_list = []
for i in range(2, 11):
    file_name = "real_cora_IID_1.0_Block_federated_" + str(i) + "layer_GCN_iter_300_epoch_3_device_num_7"
    train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = get_data(file_name)
    test_acc_list.append(test_acc)
    train_acc_list.append(list(train_acc.values())[-1])
plt.plot(list(range(2, len(train_acc_list) + 2)), train_acc_list, 'o-', label = 'Train FedGCN(0-hop)', markersize=8)
plt.plot(list(range(2, len(test_acc_list) + 2)), test_acc_list, 'o-', label = 'Test FedGCN(0-hop)', markersize=8)

test_acc_list = []
train_acc_list = []
for i in range(2, 11):
    file_name = "real_cora_IID_1.0_1hop_Block_federated_" + str(i) + "layer_GCN_iter_300_epoch_3_device_num_7"
    train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = get_data(file_name)
    test_acc_list.append(test_acc)
    train_acc_list.append(list(train_acc.values())[-1])
plt.plot(list(range(2, len(train_acc_list) + 2)), train_acc_list, 'o-', label = 'Train FedGCN(1-hop)', markersize=8)
plt.plot(list(range(2, len(test_acc_list) + 2)), test_acc_list, 'o-', label = 'Test FedGCN(1-hop)', markersize=8)

test_acc_list = []
train_acc_list = []
for i in range(2, 11):
    file_name = "real_cora_IID_1.0_2hop_Block_federated_" + str(i) + "layer_GCN_iter_300_epoch_3_device_num_7"
    train_loss, train_acc, val_loss, val_acc, test_loss, test_acc = get_data(file_name)
    test_acc_list.append(test_acc)
    train_acc_list.append(list(train_acc.values())[-1])
plt.plot(list(range(2, len(train_acc_list) + 2)), train_acc_list, 'o-', label = 'Train FedGCN(2-hop)', markersize=8)
plt.plot(list(range(2, len(test_acc_list) + 2)), test_acc_list, 'o-', label = 'Test  FedGCN(2-hop)', markersize=8)


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# naming the x axis
plt.xlabel('Number of Layers, IID = 1.0', fontsize=25)
# naming the y axis
plt.ylabel('Accuracy', fontsize=25)

plt.legend(fontsize=14, frameon=True)#, loc='lower right')
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
plt.savefig("cora_train_test_fedgcn_with_layers_iid1.eps", format='eps')
```
